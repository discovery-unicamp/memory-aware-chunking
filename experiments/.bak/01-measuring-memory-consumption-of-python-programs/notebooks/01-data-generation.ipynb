Before running the component, we need to setup a few variables and utils.

```python
from datetime import datetime

from loguru import logger

TIMESTAMP = datetime.now().strftime("%Y%m%d%H%M%S")

logger.debug("Generated global variables")
logger.debug(f"TIMESTAMP: {TIMESTAMP}")
```

```python
import sys
import os

dirs_to_add = [
    os.path.abspath('../'),
]

for dir_to_add in dirs_to_add:
    if dir_to_add not in sys.path:
        sys.path.insert(0, dir_to_add)
```

Now, we need to generate synthetic seismic data to validate the data generation process.

```python
import random
from experiment.data import generate_seismic_data, load_segy, render_random_inline

OUTPUT_DIR = f"../out/01-experiment-generation-{TIMESTAMP}"
NUM_TESTS = random.randint(3, 10)

logger.info(f"Executing {str(NUM_TESTS).zfill(2)} tests")

for i in range(NUM_TESTS):
    logger.info(f"Test #{str(i + 1).zfill(2)}")
    test_filepath = generate_seismic_data(
        output_dir=OUTPUT_DIR,
        prefix=f"test-{str(i + 1).zfill(2)}",
    )
    test_data = load_segy(test_filepath)

    logger.info(f"Test experiment shape: {test_data.shape}")
    render_random_inline(test_data)
```
